# RFM Analysis  (recency, frequency, monetary) 

```python
import datetime as dt
```

```python
data['InvoiceDate'].max()
```

```python
today_date = dt.datetime(2011,12,11)
```

```python
rfm = data.groupby('CustomerID').agg({'InvoiceDate': lambda date: (today_date - date.max()).days, #Subtract today dat - last customer buy date
                                     'InvoiceNo': lambda num: num.nunique(), #nunique For every purchase
                                     'TotalPrice': lambda TotalPrice: TotalPrice.sum()}) #Total purchases

rfm
```

```python
rfm.columns= ['Recency', 'Frequency', 'Monetary']
rfm
```

```python
rfm = rfm[rfm['Monetary'] >0]
rfm
```

```python
rfm.describe().T
```

```python
rfm['recency_score'] =pd.qcut(rfm['Recency'],5, labels=[5,4,3,2,1]) #values is reversed cuz higher means older date
rfm['frequency_score'] =pd.qcut(rfm['Frequency'].rank(method='first') ,5, labels=[1,2,3,4,5])
rfm['monetary_score'] =pd.qcut(rfm['Monetary'],5, labels=[1,2,3,4,5])
```

```python
rfm.head() #note 5 mean highest score
```

```python
rfm.sample(10)
```

```python
rfm['RFM_Score']= (rfm['recency_score'].astype(str) + rfm['frequency_score'].astype(str))
rfm.head()
```

```python
rfm.sample(10)
```

Segmenting customers using RFM Score

```python
seg_map={
    r'[1-2][1-2]':'hibernating',
    r'[1-2][3-4]':'at_risk',
    r'[1-2]5':'cant_loose',
    r'3[1-2]':'about_to_sleep',
    r'33':'need_attention ',
    r'[3-4][4-5]':'loyal',
    r'41':'promising',
    r'51':'new_customer',
    r'[4-5][2-3]':'potential_loyalist',
    r'5[4-5]':'hero ',
    
}

rfm['Segment'] = rfm['RFM_Score'].replace(seg_map, regex =True)
rfm.head()
```

```python
rfm.sample(10)
```

```python
rfm[['Segment','Recency','Frequency','Monetary']].groupby('Segment').agg(['mean','count','max']).round()
```
